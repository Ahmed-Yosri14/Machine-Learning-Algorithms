```python

import pandas as pd
from LinearRegression import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

data = pd.read_csv("cancer_risk_factors.csv")
data.drop(columns = "Patient_ID",inplace = True)
X = data.drop(columns= ["Overall_Risk_Score"],inplace = False)
Y = data['Overall_Risk_Score']

```

```python
for col in X.columns:
    X[col] = X[col].fillna(X[col].mode()[0])

print(X)
```

```python
cnt = 0
X.loc[(X["Cancer_Type"] == "Breast") & (X["Gender"] == "Male"), "Gender"] = "Female"

for gender,cancer_type in zip(X["Gender"],X["Cancer_Type"]):
    if cancer_type=="Breast" and gender =="Male":
        cnt+=1
print(cnt)
```

```python
categorical = ["Cancer_Type","Gender"]

le = LabelEncoder()

for col in categorical:
    X[col] = le.fit_transform(X[col])
print(X)
```

```python

columns = ["Cancer_Type","Gender"]


def find_classes(X,column="Cancer_Type"):
    return X[column].unique().tolist()


def count(X,class_type, column="Cancer_Type"):
    return (X[column] == class_type).sum()


for acol in columns:
    classes = find_classes(X,acol)
    print(classes)
    for aclass in classes:
        print("Frequency of class '",aclass,"' is ",count(X,aclass,acol))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=100)
print(X_train.isna().sum())   # check each feature
print(Y_train.isna().sum())   # check target


```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
pred =model.predict(X_test)

print(pred)
```
